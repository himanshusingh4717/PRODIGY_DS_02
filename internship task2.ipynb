```python

import pandas as pd
import numpy as np
data = pd.read_csv("C:/Users/user/Downloads/titanic.zip")
print(data)
```

```python
data.dropna(inplace=True)
```

# Relation between the variables (i.e. gender and survival) #

```python
pd.crosstab(data["Sex"],data["Survived"])
```

# Graphicaal Understanding #

```python
data.groupby('Sex')["Survived"].mean().plot(kind="bar")
```

```python
data['Age_grp'] = pd.cut(data["Age"],bins=[0,20,40,60,80])
data.groupby("Age_grp",observed=True)["Survived"].mean().plot(kind='bar')
```
